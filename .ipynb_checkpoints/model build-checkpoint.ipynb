```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from umap import UMAP
import umap.plot
from sklearn.model_selection import GridSearchCV, train_test_split, KFold, cross_val_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import make_scorer, silhouette_score
from sklearn.naive_bayes import MultinomialNB

from pycaret.classification import setup, compare_models, evaluate_model, plot_model
import optuna
```

```python
train = pd.read_excel('Data_Train.xlsx')
test = pd.read_excel('Data_Test.xlsx')
target = train.SECTION
train = train.drop('SECTION', axis=1).copy()
train.shape
```

```python
train.head()
```

```python
target.value_counts()
```

# Feature Transformation

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train['STORY'])
```

```python
X_train.shape
```

# UMAP

```python
# Create a UMAP object
umap = UMAP()

# Define the grid of hyperparameters to search over
param_grid = {
    'n_neighbors': [5, 10, 15, 20, 25],
    'min_dist': [0.1, 0.3, 0.4, 0.5, 0.8]
}

# Perform grid search to find the best hyperparameters
grid_search = GridSearchCV(umap, param_grid, cv=5, scoring=make_scorer(silhouette_score))
grid_search.fit(X_train, target)

# Create a new UMAP object using the best hyperparameters
best_umap = UMAP(n_neighbors=grid_search.best_params_['n_neighbors'], 
                 min_dist=grid_search.best_params_['min_dist'])

# Fit the new UMAP object on the training data and transform it into a lower-dimensional representation
X_umap = best_umap.fit_transform(X_train)
```

```python
mapper = UMAP(n_components=2).fit(X_train)
umap.plot.points(mapper, labels=target)
```

```python
mapper.get_params()
```

# Base Model

```python
model = MultinomialNB()

results=[]

kf = KFold(n_splits=10)
r = cross_val_score(model, X_train, target, scoring='accuracy', cv=kf)
print(r)
results.append(r)
print(f'Accuracy: {np.mean(results)}')
```
