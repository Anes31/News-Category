```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from umap import UMAP
import umap.plot
from sklearn.model_selection import GridSearchCV, train_test_split, KFold, cross_val_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import make_scorer, silhouette_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier

from pycaret.classification import setup, compare_models, evaluate_model, plot_model
import optuna
```

```python
train = pd.read_excel('Data_Train.xlsx')
test = pd.read_excel('Data_Test.xlsx')
target = train.SECTION
train = train.drop('SECTION', axis=1).copy()
train.shape
```

```python
train.head()
```

```python
target.value_counts()
```

# Feature Transformation

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train['STORY'])
```

```python
X_train.shape
```

# UMAP

```python
mapper = UMAP(n_components=2).fit(X_train)
X = mapper.transform(X_train)
umap.plot.points(mapper, labels=target)
```

# Base Model

```python
results={}
for n in range(2,100):
    print(n)
    X = UMAP(n_components=n).fit_transform(X_train)

    model = RandomForestClassifier(random_state=0)

    kf = KFold(n_splits=10)
    r = cross_val_score(model, X, target, scoring='accuracy', cv=kf)
    results[n]=r
```

```python
results1={}
for n in range(5,101, 5):
    print(n)
    X = UMAP(n_neighbors=n).fit_transform(X_train)

    model = RandomForestClassifier(random_state=0)

    kf = KFold(n_splits=10)
    r = cross_val_score(model, X, target, scoring='accuracy', cv=kf)
    results1[n]=r
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x=results.keys(), y=[np.mean(val).round(4) for val in results.values()])
plt.show()
```

```python
sns.lineplot(x=results1.keys(), y=[np.mean(val).round(4) for val in results1.values()])
plt.show()
```
